# Products Transformation - Silver Layer

## Summary
- Purpose: Clean and normalize product metadata from Bronze for consistent product dimensions.
- Inputs: `capstone.bronze.products` Delta table
- Outputs: `capstone.silver.products` Delta table (normalized product names and categories)
- Audit: Calls `audit_log(spark, table_name, log_path)` after write to record operation metadata.

## Key Transformations
- Apply `initcap` to product names
- Drop rescue columns and enforce schema

## Usage
- Run after Bronze ingestion. Ensure audit `log_path` is configured.


```python
from capstone_pipeline.main import transform_inticap, audit_log 

table_name = "capstone.silver.products"
log_path = "/Volumes/capstone/bronze/history"
```

```python
df_products = spark.table("capstone.bronze.products")
df_products = df_products.transform(transform_inticap,"product_name") \
    .drop("_rescue")
```

```python
(df_products.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable(table_name))
```

```python
audit_log(spark, table_name, log_path)
```
